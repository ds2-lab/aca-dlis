```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from matplotlib import pyplot
import os
import itertools
import matplotlib.backends.backend_pdf

def get_max_column_num(data_file, delimiter): 
    max_count = 0   
    with open(data_file, 'r') as temp_f:
        lines = temp_f.readlines()
        
        for l in lines:
            column_count = len(l.split(delimiter)) + 1
            max_count = max(column_count, max_count)  
    return max_count

def generate_max_column_names(max_count): 
    return [i for i in range(0, max_count)]

def get_df(path):

    delimiter=','

    col_num = get_max_column_num(path, delimiter)
    col_names = generate_max_column_names(col_num)

    data = pd.read_csv(path, header=None, delimiter=delimiter, names=col_names, engine='python')
      
    return data
```

```python
w_files = [
        '/home/ubuntu/data/attack_out/out/perf_aca_greedy_whitebox.out'
]


b_files = [
            '/home/ubuntu/data/attack_out/out/perf_aca_greedy_blackbox.out'
]

g_files = [
            '/home/ubuntu/data/attack_out/out/perf_aca_greedy_graybox.out'
]

l_files = [
            '/home/ubuntu/data/attack_out/out/perf_aca_greedy_legit.out'
]
```

```python
throughput_size = 1000000
dfs = []

for f in w_files:
    unsorted_df = get_df(f)
    sorted_df = unsorted_df.sort_values(by=[0, 1, 5, 6])
    dfs.append(sorted_df)
    
w_df = pd.concat(dfs)
w_df[8] = w_df[8].div(throughput_size)

dfs = []
for f in g_files:
    unsorted_df = get_df(f)
    sorted_df = unsorted_df.sort_values(by=[0, 1, 5, 6])
    dfs.append(sorted_df)
    
g_df = pd.concat(dfs)
g_df[8] = g_df[8].div(throughput_size)

dfs = []
for f in b_files:
    unsorted_df = get_df(f)
    sorted_df = unsorted_df.sort_values(by=[0, 1, 5, 6])
    dfs.append(sorted_df)
    
b_df = pd.concat(dfs)
b_df[8] = b_df[8].div(throughput_size)

dfs = []
for f in l_files:
    unsorted_df = get_df(f)
    sorted_df = unsorted_df.sort_values(by=[0, 1, 5, 6])
    dfs.append(sorted_df)
    
l_df = pd.concat(dfs)
l_df[8] = l_df[8].div(throughput_size)
```

```python
# print(len(b_df))
```

```python
workload_size = 10000000
budgets = [0.05, 0.1, 0.15, 0.2]
consecutive_poison_inserts = [200, 400, 600]

datasets = [0 , 1 , 2, 3]
workloads = [0.5, 0.9]
workload = 0.9
titles = {0: 'longitude', 
          1: 'longlat', 
          2: 'ycsb',
          3: 'lognormal'}

workload_titles = {0.5: 'wh', 
             0.9: 'rh',}
# print(g_df)
```

```python
all_res = {}
```

```python
for dataset in datasets:
    
    dataset_res = {}
    for workload in workloads:
        
        workload_res = {}
        
        #baseline data
        baselines_mean = [l_df[(l_df[0] == dataset) & 
                               (l_df[7] == workload) & 
                               (l_df[5] == 0)][8].mean()] * len(budgets)
        baselines_std = [np.std(l_df[(l_df[0] == dataset) & 
                                     (l_df[7] == workload) & 
                                     (l_df[5] == 0)][8])] * len(budgets)
        
        workload_res['Modified_ALEX'] = (baselines_mean, baselines_std)   
        
        
        #whitebox data
        for c in consecutive_poison_inserts:
            whitebox_mean = []
            whitebox_std = []
            for b in budgets:
                
                
        
                data = w_df[(w_df[0] == dataset) 
                                   & (w_df[7] == workload) 
                                   & (w_df[5] == b) & (w_df[6] == c)][8].to_numpy()
        
                whitebox_mean.append(data.mean())
                whitebox_std.append(np.std(data))
        
            
            workload_res['White-box'+str(c)] = (whitebox_mean, whitebox_std)
            
            
        #blackbox data
        for c in consecutive_poison_inserts:
            blackbox_mean = []
            blackbox_std = []
            for b in budgets:

                data = b_df[(b_df[0] == dataset) 
                                   & (b_df[7] == workload) 
                                   & (b_df[5] == b) & (b_df[6] == c)][8].to_numpy()

                blackbox_mean.append(data.mean())
                blackbox_std.append(np.std(data))

            workload_res['Black-box'+str(c)] = (blackbox_mean, blackbox_std)
            
        #graybox data
        for c in consecutive_poison_inserts:
            graybox_mean = []
            graybox_std = []
            for b in budgets:

                data = g_df[(g_df[0] == dataset) 
                                   & (g_df[7] == workload) 
                                   & (g_df[5] == b) & (g_df[6] == c)][8].to_numpy()

                graybox_mean.append(data.mean())
                graybox_std.append(np.std(data))

            workload_res['Gray-box'+str(c)] = (graybox_mean, graybox_std)
        
        
        dataset_res[workload] = workload_res
                        
    
    all_res[dataset] = dataset_res
```

```python
print(all_res)
```

```python
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['font.size']=20
w_x = [ b * 100 * 0.5 for b in budgets]
r_x = [ b * 100 * 0.1 for b in budgets]

x = [ b * 100 for b in budgets]

# figsize=(6,8)

fig, axes = plt.subplots(2,1, constrained_layout=False, figsize=(6,10))

ls=['-','--','-.']
# cs=['blue', 
#     'black', 'dimgray', 'silver', 
#     'lightslategray','lightslategrey', 'slategrey', 
#     'crimson', 'pink', 'lightpink']
markers=["d", "v", "s", "X", ">", "<", "8", "^", "2"]
markersize= 12
# plt.figure(figsize=(6,5))
axes[0].grid(axis='both',which='both', zorder=0);
axes[1].grid(axis='both',which='both', zorder=0);
axes[0].set_xticks(w_x)
capsize=6
linewidth=3

dataset_idx=3
dataset_res = all_res[dataset_idx]


workload = 0.5
wh_data = dataset_res[workload]

axes[0].set_ylabel('Throughput (M ops/s)')
axes[0].set_xlabel('Percentage (% total requests)')
axes[0].set_title('Write-heavy')

cnt=0
for it, measurement in wh_data.items():
    axes[0].errorbar(w_x, measurement[0], 
                        yerr=measurement[1],
                 zorder=3,
#                       color = cs[cnt%len(cs)],
                     markersize=markersize,
#                      markerfacecolor="None",
#                      markeredgecolor=ls[cnt % len(ls)],
                        label=it,
                      capsize=capsize,
                     linewidth=linewidth,
#                      elinewidth = elinewidth,
                     ls=ls[cnt % len(ls)],
                    marker=markers[cnt%len(markers)])
    cnt+=1
    

workload = 0.9
rh_data = dataset_res[workload]

axes[1].set_ylabel('Throughput (M ops/s)')
axes[1].set_xlabel('Percentage (% total requests)')
axes[1].set_title('Read-heavy')
axes[1].set_xticks(r_x)
cnt=0
for it, measurement in rh_data.items():
    axes[1].errorbar(r_x, measurement[0], 
                        yerr=measurement[1],
                 zorder=3,
#                       color = cs[cnt%len(cs)],
                     markersize=markersize,
#                      markerfacecolor="None",
#                      markeredgecolor=ls[cnt % len(ls)],
                        label=it,
                      capsize=capsize,
                     linewidth = linewidth,
                     ls=ls[cnt % len(ls)],
                    marker=markers[cnt%len(markers)])
    cnt+=1

    
handles, labels = axes[0].get_legend_handles_labels()
print(len(labels))

if dataset_idx == 0:
    fig.legend(ncol=2, 
               columnspacing=0.2, 
               handletextpad=0.5, 
               loc='lower center', bbox_to_anchor=(0.6, 0.61),
               labels=labels,
              fontsize=15)
    

plt.tight_layout(pad=0.4,w_pad=0.5,h_pad=1.0)          
plt.savefig(fname='ac_time_' + titles[dataset_idx] + '.pdf',format="pdf")
    

```
